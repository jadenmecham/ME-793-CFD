```python
import sympy as sp
```

```python
fj, fjm1, fjp1, fjm2, fjp2  = sp.symbols(r'f_j f_{j-1} f_{j+1} f_{j-2} f_{j+2}')
df, d2f, d3f, d4f, d5f, d6f = sp.symbols(r'f^{(1)} f^{(2)} f^{(3)} f^{(4)} f^{(5)} f^{(6)}')
x, h  = sp.symbols(r'x \Delta')
```

## Problem 1

Define the 6th order Taylor Series expansion about point $x_j$ for a uniform mesh with spacing $\Delta$. 

```python
t6 = fj + df*h + d2f*h**2/2 + d3f*h**3/6 + d4f*h**4/24 + d5f*h**5/120 + d6f*h**6/720
t6
```

For a 5 point central difference, we must expand $f(x\pm h)$ and $f(x\pm 2h)$.

```python
fxph = t6 - fjp1
fxmh = t6.subs(h, -h) - fjm1
fxp2h = t6.subs(h, 2*h) - fjp2
fxm2h = t6.subs(h, -2*h) - fjm2
```

```python
fxph
```

```python
fxmh
```

```python
fxp2h
```

```python
fxm2h
```

Now, sum symmetric terms. This eliminates odd order derivatives. 

```python
s1 = fxph + fxmh
s2 = fxp2h + fxm2h
```

```python
s1
```

```python
s2
```

To isolate $f^{(4)}_j$, we need to eliminate the $f^{(2)}_j$ term by performing $S_2-4S_1$.

```python
s = s2 - 4*s1
s
```

Rearrange to solve for $f_j^{(4)}$.

```python
d4f_sol = sp.solve(s, d4f)[0]
d4f_sol
```

## Problem 2

For a 5 point central difference, we must expand $f(x\pm h)$ and $f(x\pm 2h)$. In this case, we expand to $\pm \frac{1}{2}\Delta$ and $\pm \frac{3}{2}\Delta$.

```python
fj, fjm1, fjp1, fjm2, fjp2  = sp.symbols(r'f_j f_{j-\frac{1}{2}} f_{j+\frac{1}{2}} f_{j-\frac{3}{2}} f_{j+\frac{3}{2}}')
```

```python
fxph = t6.subs(h, (h/2)) - fjp1
fxmh = t6.subs(h, (-h/2)) - fjm1
fxp2h = t6.subs(h, (3*h/2)) - fjp2
fxm2h = t6.subs(h, (-3*h/2)) - fjm2
```

Subtract symmetric terms to eliminate even order derivatives. 

```python
s1 = fxph - fxmh
s1
```

```python
s2 = fxp2h - fxm2h
s2
```

Solve $S_1$ for $f^{3}$

```python
df3_sol = sp.solve(s1, d3f)[0]
df3_sol
```

Plug $f^{(3)}$ into $S_2$.

```python
s2 = s2.subs(d3f, df3_sol)
s2
```

Now solve $S_2$ for $f^{(1)}$.

```python
f1_sol = sp.solve(s2, df)[0].expand()
f1_sol
```

```python
fxph
```

```python
f1 = sp.solve(f_sol, df)[0].expand()
f1
```
